## How does tenure affect customer churn?

```python
short_range = (0, 12)         # up to 12 months
intermediate_range = (13, 36)  # 13 to 36 months
long_range = (37, float('inf'))  # 37 months and above

#categorize tenure periods
def categorize_tenure(tenure):
    if tenure >= long_range[0]:
        return 'Long Period'
    elif tenure >= intermediate_range[0]:
        return 'Intermediate Period'
    elif tenure >= short_range[0]:
        return 'Short Period'
    else:
        return 'Unknown'
df['tenure_category'] = df['tenure'].apply(categorize_tenure)
```

```python
df.head()
```

```python
df['tenure'].mean()
```

The average tenure of a customer is 32.37 months. This means that the average customer has been with the company for 32.37 months.

```python
df.groupby('tenure_category')['Churn'].value_counts(normalize=True).mul(100).rename('percentage').reset_index()
```

- **For the "Intermediate Period," the churn percentage is around 25.54%, meaning that approximately 25.54% of customers in this category have churned.**
- **For the "Long Period," the churn percentage is lower at 11.93%.**
- **In the "Short Period" category, the churn percentage is relatively high at 47.44%.**
- **It seems that customers with longer tenures have lower churn percentages compared to those with shorter tenures.**

<br>
- **Customers in the "Long Period" category seem to have a higher likelihood of staying with the service, as indicated by the lower churn percentage.**
